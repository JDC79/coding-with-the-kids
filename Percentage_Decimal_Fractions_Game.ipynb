```
#@title
name = input("What is your name? ")
print ("\u001b[36m Hello " + name + ", Welcome to Percentage Decimal Fraction - the game to test your Mathematical knowledge!")
print ("\u001b[36m You will answer a set of questions, be careful with your answers ...... let's begin!")
question_1 = int(input("Question 1 - What is 25% of 200? "))
if question_1 == 50:
  print("\u001b[32m CORRECT! 1 point to you")
else:
  print("\u001b[31m Uh-oh ... WRONG! The correct answer is 50.")
question_2 = int(input("\u001b[36m Question 2 - What is 10% of 1,500? "))
if question_2 == 150:
  print("\u001b[32m AMAZING! another point to you")
else:
  print("\u001b[31m ERROR! The correct answer is 150.")
question_3 = int(input("\u001b[36m Question 3 - What is 1/2 of 350? "))
if question_3 == 175:
  print("\u001b[32m BRILLIANT! 1 more point. Are you keeping score? ")
else:
  print("\u001b[31m Too bad .... that ones INCORRECT! The correct answer is 175.")
question_4 = int(input("\u001b[36m Question 4 - What is 1/3 of 24? "))
if question_4 == 8:
  print("\u001b[32m FANTASTIC! give yourself 1 point")
else:
  print("\u001b[31m Sorry .... that WASNT RIGHT! The correct answer is 8.")
question_5 = int(input("\u001b[36m Question 5 - What is 2/5 of 100? "))
if question_5 == 40:
  print("\u001b[32m SPOT ON! thats another point for you")
else:
  print("\u001b[31m TRY AGAIN! The correct answer is 40.")
question_6 = int(input("\u001b[36m Question 6 - What is 1/4 of 780 is? "))
if question_6 == 195:
  print("\u001b[32m BOOM! 1 point to you. How many points do you have now?")
else:
  print("\u001b[31m Oooops....ZERO POINTS! The correct answer is 195.")
question_7 = float(input("\u001b[36m Question 7 - Write two tenths as decimal. "))
if question_7 == 0.2:
  print("\u001b[32m TOP JOB! another point to you")
else:
  print("\u001b[31m SORRY, IT'S NOT THAT ANSWER. The correct answer is 0.2 ")
question_8 = float(input("\u001b[36m Question 8 - Write eight tenths as a decimal. "))
if question_8 == 0.8:
  print("\u001b[32m TERRIFIC! 1 more point")
else:
  print("\u001b[31m BETTER LUCK NEXT TIME! The correct answer is 0.8 ")
question_9 = int(input("\u001b[36m Question 9 - What is 10% of 3,000? "))
if question_9 == 300:
  print("\u001b[32m GREAT WORK! give yourself 1 point")
else:
  print("\u001b[31m ONE POINT TO ME! The correct answer is 300. ")
question_10 = int(input("\u001b[36m Question 10 - What is 75% of 1,000,000? (don't enter any commas between your numbers) "))
if question_10 == 750000:
  print("\u001b[32m REMARKABLE! thats another point for you")
else:
  print("\u001b[31m Hmmmmn .... THINK AGAIN! The correct answer is 750000.") 
score = int(input("That's the end of the quiz. How many points did you score? Add up your GREEN answers and enter 0-10 "))
if score== 10:
  print("Fantastic!! Well done, you scored 10 out of 10 so thats 100%")
elif score>= 8:
  print("Great work, you scored over 75%")
elif score>= 5:
  print("Good effort, you scored half the points available so thats over 50%")
else:
  print("This time you scored below 50%, why not take the quiz again and see if you can do better!")
print("\u001b[36m Thanks for taking part. Have a good day :)")

import PIL
from PIL import Image
file_image = ("rewards.png")
Image.open(file_image)
```
